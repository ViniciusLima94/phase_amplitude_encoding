### Spectral analysis 

Compute the single-trial and time-resolved CSD using Multitapers or Morlet wavelet

```python
# Select data
data = data_chain_k
trials = cs_chain_k
trials = trials.flatten()
times = np.arange(0, data.shape[2], 1)
roi = np.arange(0, data.shape[1], 1)
```

```python
# Params
win = 0.2
step = 0.01
freqs = np.arange(10, 70, 2)
sfreq = 600

# Spectral matrix S(n_epochs, n_chans, n_freqs, n_tapers, n_times)
# out = tfr_array_multitaper(data, sfreq=sfreq, freqs=freqs, 
#                          time_bandwidth=5, use_fft=True, decim=1, output='complex', verbose=None)

# Spectral matrix S(n_epochs, n_chans, n_freqs, n_times)
out = tfr_array_morlet(data, sfreq=sfreq, freqs=freqs, 
                       n_cycles=freqs/7, use_fft=True, decim=1, output='complex', verbose=None)

```

```python
# xarray 
S = xr.DataArray(out, dims=('trials', 'roi', 'freqs', 'times'),
                 coords=(trials, roi, freqs, times))
```

### Phase relation GMCI analysis

Across-trials analyses using Frequency domain Gaussian-copula Mutual Information (GCMIF)

```python
from hoi.core import get_mi

# Define the function to compute MI using HOI and JAX
mi_fcn = get_mi("gc")

# vectorize the function to first and second dimension
gcmi = jax.vmap(jax.vmap(mi_fcn, in_axes=0), in_axes=0)

# Select data for nodes
x = S.sel(roi=[0]).data.squeeze()
y = S.sel(roi=[1]).data.squeeze()
z = S.sel(roi=[2]).data.squeeze()

# Edge activity (with and without normalisation)
e1 = x * np.conj(y)
e2 = y * np.conj(z)
e3 = x * np.conj(z)

# Real and Imag parts pf edge activity
e1r, e1i = np.real(e1), np.imag(e1)
e2r, e2i = np.real(e2), np.imag(e2)
e3r, e3i = np.real(e3), np.imag(e3)

# Stack complex values
E1 = np.stack((e1r, e1i), axis=1)   
E2 = np.stack((e2r, e2i), axis=1)
E3 = np.stack((e3r, e3i), axis=1)
E12 = np.stack((e1r, e1i, e2r, e2i), axis=1)
E23 = np.stack((e2r, e2i, e3r, e3i), axis=1)
E123 = np.stack((e1r, e1i, e2r, e2i, e3r, e3i), axis=1)

# Swap axis 0 with 3 and 1 with 2
E1 = np.moveaxis(E1, [0, 1], [-1, -2])
E2 = np.moveaxis(E2, [0, 1], [-1, -2])
E3 = np.moveaxis(E3, [0, 1], [-1, -2])
E12 = np.moveaxis(E12, [0, 1], [-1, -2])
E23 = np.moveaxis(E23, [0, 1], [-1, -2])
E123 = np.moveaxis(E123, [0, 1], [-1, -2])

# Stims across trials
stim = trials
stim = np.expand_dims(stim, axis=(0, 1))
stim = np.tile(stim, (len(freqs), len(times), 1, 1))

# # MI frequency domain
# mi_freq[i, :] = gcmi_nd_cc(X, Y, mvaxis=1, traxis=0)

# # Coherence
# coh[i, :] = np.abs(e1.mean(axis=0)).squeeze()

# # Phase-Locking Value
# plv[i, :] = np.abs(e1n.mean(axis=0)).squeeze()

# Copnorm
E1 = copnorm_nd(E1, axis=-1)
E2 = copnorm_nd(E2, axis=-1)
E3 = copnorm_nd(E3, axis=-1)
E12 = copnorm_nd(E12, axis=-1)
E23 = copnorm_nd(E23, axis=-1)
E123 = copnorm_nd(E123, axis=-1)
stim = copnorm_nd(stim, axis=-1)

# MIF for edge encoding
mif_e1 = gcmi(E1, stim).T
mif_e2 = gcmi(E2, stim).T
mif_e3 = gcmi(E3, stim).T
mif_e12 = gcmi(E12, stim).T
mif_e23 = gcmi(E23, stim).T
mif_e123 = gcmi(E123, stim).T

red = np.minimum(mif_e1, mif_e2)

syn = mif_e123 - np.maximum(mif_e1, mif_e2)
```

### Task-related Frequency-domain Edge Encoding

MIF(Edge; Stim) where Edge is the phase difference

```python
# Init
mif_e1 = np.zeros((len(times), len(freqs)))
mif_e12 = np.zeros((len(times), len(freqs)))
mif_e21 = np.zeros((len(times), len(freqs)))
mif_e11 = np.zeros((len(times), len(freqs)))
mif_e22 = np.zeros((len(times), len(freqs)))

# Stims across trials
stim = trials

# Single trial and time-resolved analysis
for i in times[100:-25]: 

    # Select data for nodes
    s = S.sel(times=[i])
    x = s.sel(roi=[0]).data
    y = s.sel(roi=[1]).data

    # Select past data for nodes
    s = S.sel(times=[i-75])
    xp = s.sel(roi=[0]).data
    yp = s.sel(roi=[1]).data

    # Can add the option to normalise the phase difference to have plv
    x = x / np.abs(x)
    y = y / np.abs(y)
    xp = xp / np.abs(xp)
    yp = yp / np.abs(yp)
    
    # Instantaneous Edge activity phase difference
    e1 = x * np.conj(y)

    # Lagged Edge activity phase difference
    e12 = x * np.conj(yp)
    e21 = y * np.conj(xp)
    e11 = x * np.conj(xp)
    e22 = y * np.conj(yp)

    # Real and Imag parts of instantaneous phase difference
    e1r = np.real(e1)
    e1i = np.imag(e1) 

    # Real and Imag parts of lagged phase difference
    e12r = np.real(e12)
    e12i = np.imag(e12)
    e21r = np.real(e21)
    e21i = np.imag(e21)
    e11r = np.real(e11)
    e11i = np.imag(e11)
    e22r = np.real(e22)
    e22i = np.imag(e22)
    
    # Stack complex values
    E1 = np.stack((e1r, e1i), axis=1) 
    E12 = np.stack((e12r, e12i), axis=1)
    E21 = np.stack((e21r, e21i), axis=1)
    E11 = np.stack((e11r, e11i), axis=1)
    E22 = np.stack((e22r, e22i), axis=1)

    # Remove 3rd dim
    E1 = np.squeeze(E1)
    E12 = np.squeeze(E12)
    E21 = np.squeeze(E21)
    E11 = np.squeeze(E11)
    E22 = np.squeeze(E22)

    # Swap data first and last dim
    E1 = np.swapaxes(E1, 0, 2)
    E12 = np.swapaxes(E12, 0, 2)
    E21 = np.swapaxes(E21, 0, 2)
    E11 = np.swapaxes(E11, 0, 2)
    E22 = np.swapaxes(E22, 0, 2)

    # MIF with stim for local encoding at the edge level
    for f in range(len(freqs)):
        # MIF for edge encoding
        mif_e1[i, f] = gcmi_1d_cc(E1[f,:,:].squeeze(), stim)
        mif_e12[i, f] = gcmi_1d_cc(E12[f,:,:].squeeze(), stim)
        mif_e21[i, f] = gcmi_1d_cc(E21[f,:,:].squeeze(), stim)
        mif_e11[i, f] = gcmi_1d_cc(E11[f,:,:].squeeze(), stim)
        mif_e22[i, f] = gcmi_1d_cc(E22[f,:,:].squeeze(), stim)

red = np.minimum(mif_e11, mif_e22)

unique_11 = mif_e11 - red

unique_22 = mif_e22 - red


red = np.minimum(mif_e12, mif_e21)

unique_12 = mif_e12 - red

unique_21 = mif_e21 - red

red = np.minimum(mif_e12, mif_e22)

unique_12_22 = mif_e12 - red

red = np.minimum(mif_e21, mif_e11)

unique_21_11 = mif_e21 - red

```

```python
import xarray as xr
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(14, 4))
# Plot time-frequency MI
mi = xr.DataArray(mif_e1, dims=('times', 'freqs'), coords=(times,freqs))
# Plot mi as image
mi.plot(x='times', y='freqs', cmap='viridis', vmin=0, ax=axs[0])
axs[0].set_title("MIF with stim for local encoding at the edge level")
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Frequency (Hz)')
# Plot the frequency averaged MI
axs[1].plot(times, mif_e1.mean(axis=1))
axs[1].set_title('MIF with stim for local encoding at the edge level')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('MI (bits)')
# add grid
axs[1].grid()
# tight xlim
axs[1].set_xlim([times[0], times[-1]])
plt.show()
```
