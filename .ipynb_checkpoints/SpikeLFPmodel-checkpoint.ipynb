```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torchvision
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
```

```python
class SLM(torch.nn.Module):

    def __init__(
        self,
        f: float,
        vr: float,
        vth: float,
        tau: float,
        eta_ud: float,
        eta_bu: float,
        beta: float,
        h: float,
    ):
        super().__init__()

        # LFP parameters
        self.a = -5
        self.beta = beta * torch.sqrt(torch.tensor(h))
        self.f = f

        # Spike parameters
        self.vr = vr
        self.vth = vth
        self.tau = tau

        # LFP-spike connection parameterds
        self.eta_ud = eta_ud
        self.eta_bu = eta_bu

        # Simulation parameters
        self.h = h

    def __ode_lfp(self, z: torch.Tensor, a: torch.Tensor, w: torch.Tensor):
        return z * (a + 1j * w - torch.abs(z) ** 2)

    def __ode_spikes(self, v: torch.Tensor, vr: torch.Tensor, tau: torch.Tensor):
        return -(v - vr) / tau

    def loop(self, z_t: torch.Tensor, v_t: torch.Tensor, x_t):

        # integrate lfp
        z_t = (
            z_t
            + self.h * self.__ode_lfp(z_t, self.a, self.f)
            + self.beta * torch.randn(z_t.shape)
            + self.beta * 1j * torch.randn(z_t.shape)
            + self.h * self.eta_bu * x_t
        )

        # integrate spikes
        v_t = (
            v_t
            + self.h * self.__ode_spikes(v_t, self.vr, self.tau)
            + self.h * self.eta_ud * torch.real(z_t)
        )

        x_t = v_t >= self.vth

        if x_t:
            v_t = self.vr

        return z_t, v_t, x_t

    def forward(self, n_times: int):

        rec_z_t = torch.zeros(n_times, dtype=torch.complex64)
        rec_v_t = torch.zeros(n_times, dtype=torch.float32)
        rec_x_t = torch.zeros(n_times, dtype=torch.int)

        z_t = self.h * torch.randn(1) + 1j * self.h * torch.randn(1)
        v_t = torch.tensor(-70)
        x_t = torch.tensor(0)

        for t in range(n_times):
            z_t, v_t, x_t = self.loop(z_t, v_t, x_t)
            rec_z_t[t] = z_t
            rec_v_t[t] = v_t
            rec_x_t[t] = x_t

        return rec_z_t, rec_v_t, rec_x_t
```

```python
model = SLM(
        40,
        -70,
        -60,
        20,
        100000,
        10,
        1e-4,
        1e-4)
```

```python
out = model(10000)
```

```python
plt.subplot(311)
plt.plot(out[0].real)
plt.subplot(312)
plt.plot(out[1].real)
plt.subplot(313)
plt.plot(out[2].real)
plt.tight_layout()
```
